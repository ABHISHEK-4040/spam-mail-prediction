```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
import pickle
```

```python

mail_data=pd.read_csv("mail_data.csv")
```

```python
mail_data.head()
```

```python
mail_data.tail()
```

```python
mail_data.shape
```

```python
mail_data.isnull().sum()
```

```python
mail_data['Category'].value_counts()
```

```python
labelencoder=LabelEncoder()
```

```python
mail_data['Category']=labelencoder.fit_transform(mail_data['Category'])
```

```python
mail_data.head()
```

0=ham
1=spam

```python
a=mail_data[mail_data['Category']==0]
b=mail_data[mail_data['Category']==1]
```

```python
a=a.sample(n=747)
```

```python
mail_data=pd.concat([a,b],ignore_index=True)
```

```python
mail_data['Category'].value_counts()
```

```python
x=mail_data['Message']
y=mail_data['Category']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=3)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
vectorizer=TfidfVectorizer(min_df=1,stop_words="english",lowercase=True)
```

```python
vectorizer.fit(x_train)
```

```python
x_train=vectorizer.transform(x_train)
```

```python
print(x_train)
```

```python
x_test=vectorizer.transform(x_test)
```

```python
print(x_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
```

```python
x_train_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print(f"X Train Accuracy: {x_train_accuracy*100:.1f}%")
```

```python
x_test_prediction=model.predict(x_test)
```

```python
x_test_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print(f"X Test Accuracy: {x_test_accuracy*100:.1f}%")
```
